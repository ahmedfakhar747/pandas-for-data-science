# 📌 1. Introduction

Data preprocessing and **Exploratory Data Analysis (EDA)** are fundamental steps in every data science and machine learning workflow. Before building any model, it is essential to understand the dataset, identify data quality issues, clean inconsistencies, and explore meaningful patterns.

In this project, we will use the **Titanic** dataset to perform a complete preprocessing and EDA workflow using **Pandas**. The notebook covers data exploration, handling missing values, removing duplicate records, feature engineering, statistical analysis, and data visualization.

By the end of this project, the dataset will be **clean, well-understood, and ready for machine learning**, while also revealing valuable insights into the factors that influenced passenger survival.
