# 🧹 6. Data Cleaning

Raw datasets often contain ***inconsistencies*** such as missing values, duplicate records, or incorrect data types. In this section, we ***clean*** and ***prepare*** the dataset to ensure it is ***accurate***, ***consistent***, and ready for analysis.

## 6.1 Handle Missing Values

Missing values can ***affect the quality*** of analysis and machine learning models. Here, we identify missing data and apply ***appropriate techniques*** to handle it while preserving the ***integrity*** of the dataset.


```python
df.Age = df.Age.fillna(df.Age.median())
```

> **We use the median for the `Age` column because it is less affected by outliers and provides a more reliable estimate than the mean.**

```python
df.Cabin = df.Cabin.fillna("Unknown")
```

> **Since the `Cabin` column contains a large number of missing values, we replace them with `"Unknown"` (or drop the column) instead of using the mode to avoid introducing bias.**

```python
df.Embarked = df.Embarked.fillna(df.Embarked.mode()[0])
```

> We use the **mode** for the `Embarked` column because it is **categorical data**, and the mode preserves the most frequently occurring category.

```python
df
```

```python
df.isnull().sum()
```

Now the data is fully ***inconsistent*** and has no `NaN` values.
