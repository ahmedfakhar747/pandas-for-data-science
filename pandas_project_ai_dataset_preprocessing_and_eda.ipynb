# 🧹 6. Data Cleaning

Raw datasets often contain ***inconsistencies*** such as missing values, duplicate records, or incorrect data types. In this section, we ***clean*** and ***prepare*** the dataset to ensure it is ***accurate***, ***consistent***, and ready for analysis.

## 6.1 Handle Missing Values

Missing values can ***affect the quality*** of analysis and machine learning models. Here, we identify missing data and apply ***appropriate techniques*** to handle it while preserving the ***integrity*** of the dataset.


```python
df.Age = df.Age.fillna(df.Age.median())
```

> **We use the median for the `Age` column because it is less affected by outliers and provides a more reliable estimate than the mean.**

```python
df.Cabin = df.Cabin.fillna("Unknown")
```

> **Since the `Cabin` column contains a large number of missing values, we replace them with `"Unknown"` (or drop the column) instead of using the mode to avoid introducing bias.**

```python
df.Embarked = df.Embarked.fillna(df.Embarked.mode()[0])
```

> We use the **mode** for the `Embarked` column because it is **categorical data**, and the mode preserves the most frequently occurring category.

```python
df
```

```python
df.isnull().sum()
```

Now the data is fully ***inconsistent*** and has no `NaN` values.

Clear and consistent column names ***improve readability*** and make the dataset easier to work with. ***If necessary***, we rename columns using ***meaningful and standardized*** names.

# ⚙️ 7. Feature Engineering

## 7.1 Create New Features
Feature engineering involves creating ***new variables*** from existing data to uncover additional information. These new features can ***improve data analysis*** and often ***enhance the performance*** of future machine learning models.

A passenger's ***family size*** is calculated using:

```python
df["Family_Size"] = df["Siblings_Spouses"] + df["Parents_Children"] + 1
df
```

Checking whether passenger ***is alone*** or not:

```python
df["Is_Alone"] = (df["Family_Size"]==1).astype(int)
df.head()
```

## 7.2 Categorize Age Groups
Instead of working with individual ages, we group passengers into meaningful ***age categories*** (such as Child, Teen, Adult, and Senior). This makes the data ***easier to analyze*** and helps identify treds across differnt age groups.

```python
bins=[0, 12, 19, 35, 60, 100]
labels=["Child", "Teen", "Adult", "Middle Aged", "Senior"]

df["Age_Group"] = pd.cut(df.Age, bins=bins, labels=labels)
df
```

## 7.3 Extract Passenger Titles from Name

Passenger names contain titles such as ***Mr., Mrs., Miss., Master,*** and others. Extracting these titles provides ***additional demographic information*** that can reveal social patterns and serve as a valuable feature for analysis and machine learning.

```python
df["Title"] = df.Name.str.extract(r",\s*([^\.]+)\.")
df
```

# 📊 8. Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) helps us understand the dataset by ***uncovering patterns, trends, and relationships*** between variables. Through ***statistical summaries and visualizations***, we can gain meaningful insights that support better decision-making and prepare the data for future machine learning tasks.

## 8.1 Survival Analysis

Analyze ***the factors*** that influenced passenger survival on the Titanic by ***comparing survival rates*** across different groups and identifying key patterns.

### 8.1.1 Overall Survival Rate
Examine the percentage of passengers who survived versus those who did not.

```python
per_survived = df.Survived.value_counts(normalize=True) * 100
per_survived
```

```python
per_survived.index = ["Not Survived", "Survived"]
per_survived.plot.bar(
     xlabel="Survival Status",
     ylabel="Percentage (%)",
     title="Overall Survival Rate"
)
```

### 8.1.2 Survival by Gender
Compare survival rates between male and female passengers.

```python
gen_survived = (
    df.groupby("Gender").Survived
      .value_counts()
      .unstack()
)

gen_survived.columns = ["Did Not Survive", "Survived"]

gen_survived.plot.bar(
    title="Passenger Survival Count by Gender",
    xlabel="Gender",
    ylabel="Number of Passengers",
)
```

### 8.1.3 Survival by Passenger Class
Investigate how ticket class affected the likelihood of survival.

```python
pclass_survived = (df.groupby("Passenger_Class").Survived.value_counts().unstack())

pclass_survived.columns = ["Did Not Survive", "Survived"]

pclass_survived.plot.bar(
    title="Passenger Survival Count by Class",
    xlabel="Class",
    ylabel="Number of Passengers",
)
```

### 8.1.4 Survival by Age Group
Analyze survival trends across different age categories.

```python
age_survived = (df.groupby("Age_Group").Survived.value_counts().unstack())

age_survived.columns = ["Did Not Survive", "Survived"]

age_survived.plot.bar(
    title="Passenger Survival Count by Age Group",
    xlabel="Age Group",
    ylabel="Number of Passengers",
)
```

### 8.1.5 Survival by Embarkation Port
Explore whether the port where passengers boarded influenced survival outcomes.

```python
emb_survived = (df.groupby("Embarked").Survived.value_counts().unstack())

emb_survived.columns = ["Did Not Survive", "Survived"]

emb_survived.plot.bar(
    title="Passenger Survival Count by Embarkation Port",
    xlabel="Embarkation Port",
    ylabel="Number of Passengers",
)
```
