# 📊 Level 1: What is Data Visualization?

**Data Visualization** is the graphical representation of data using charts, graphs, and other visual elements. It helps transform raw data into meaningful insights, making it easier to understand patterns, trends, relationships, and outliers.

## 1.1 Why is it Important?

- 📈 Identifies trends and patterns
- 🔍 Reveals relationships between variables
- 🚨 Detects outliers and anomalies
- 💡 Simplifies data analysis and decision-making

> **A good visualization turns complex data into clear insights.** 📊

## 1.2 Benefits of Data Visualization

- Makes complex data easier to understand
- Reveals hidden patterns and trends
- Supports data-driven decision making
- Simplifies comparisons between categories
- Helps identify errors and outliers
- Improves communication with stakeholders

## 1.3 Common Visualization Techniques

Some of the most widely used visualization methods include:

- 📈 Line Charts
- 📊 Bar Charts

- 📉 Histograms
- 🔥 Heatmaps
- ⚪ Scatter Plots
- 📦 Box Plots
- 🗺️ Geographic Maps

Each visualization type is designed to answer different analytical questions.

# 🌈 Level 3: Understanding Style Sheets

A **Style Sheet** is a collection of predefined visual settings that control the appearance of plots and charts. Instead of customizing colors, fonts, gridlines, and backgrounds manually, you can apply a style sheet to instantly change the overall look of your visualizations.

Some popular styles include:

* `ggplot`
* `seaborn-v0_8`
* `dark_background`
* `fivethirtyeight`
* `classic`

## 3.1 Benefits of Style Sheets

- 🎨 Improves the appearance of charts
- ⚡ Applies consistent styling with a single line of code
- 📊 Makes visualizations more professional and readable

> **Style sheets help create visually appealing and consistent plots with minimal effort.**

# 📈 Level 4: Drawing Area Plot

An **area plot** is similar to a **line plot**, but the area beneath the line is filled with color. It is used to show **trends over time** while emphasizing the **magnitude** of the values.

**Best used for:**

* 📊 Sales trends
* 👥 Population growth
* 🌐 Website traffic
* 💰 Revenue over time

**Key Point:**

* **Line Plot:** Shows trends.
* **Area Plot:** Shows trends **and** the magnitude (volume) of the data.

## 4.1 DataFrame Creation

```python
df = pd.DataFrame(
    np.random.rand(10, 4),
    columns=["a", "b", "c", "d"]
)
df
```

## 4.2 Area Plot Creation

```python
df.plot.area()
```

For unstacked area plot:

```python
df.plot.area(stacked=False)
```

For only one column area plot:

```python
df.plot.area(y="a")
```

```python
df.plot.area(x="a")
```

# 📊 Level 5: Demonstrating Bar Plot

## 5.1 Theory of Bar Plot

A **bar plot (bar chart)** uses **rectangular bars** to compare values across different categories. The **height (or length)** of each bar represents the value.

**Best used for:**

* 📚 Student marks
* 💰 Product sales
* 🌍 Population by country
* 🛒 Category-wise comparisons

**Key Point:**

* **Bar Plot:** Compares values between categories.
* **Area Plot:** Shows trends and magnitude over time.

## 5.2 Implementation

```python
df.plot.bar()
```

To have a stacked bar we have to `stacked=True`

```python
df.plot.bar(stacked=True)
```

# 📈 Level 6: Visualizing Line Plot

## 6.1 Theory of Line Plot

A **line plot (line chart)** uses **points connected by lines** to show how data changes over time or across an ordered sequence. It is ideal for visualizing **trends and patterns**.

**Best used for:**

* 🌐 Website traffic

**Key Point:**

* **Line Plot:** Shows trends over time.
* **Bar Plot:** Compares categories.
* **Area Plot:** Shows trends while emphasizing the magnitude of values.

## 6.2 Implementation

```python
df.plot.line()
```

# 🔵 Level 7: Drawing Scatter Plot

## 7.1 Theory of Scatter Plots

A **scatter plot** displays **individual data points** to show the relationship between **two numerical variables**. It helps identify **correlations, clusters, and outliers**.

**Best used for:**

* 📚 Study hours vs. Exam scores
* 💰 Advertising spend vs. Sales
* 📏 Height vs. Weight
* 🌡️ Temperature vs. Ice cream sales

**Key Point:**

* **Scatter Plot:** Shows the relationship between two numerical variables.
* **Line Plot:** Shows trends over time.
* **Bar Plot:** Compares categories.
* **Area Plot:** Shows trends with emphasis on magnitude.

## 7.2 Making DataFrame

```python
students = pd.DataFrame({
    "Study Hours": [1, 2, 2.5, 3, 4, 5, 5.5, 6, 7, 8],
    "Exam Score":  [45, 50, 58, 60, 68, 74, 78, 85, 90, 96]
})

students
```

## 7.2 Implementation

```python
students.plot.scatter(
    x="Study Hours",
    y="Exam Score",
    s=100,
    title="Study Hours vs Exam Score"
)
```

## 8.1 Theory of Box Plots

A **box plot (box-and-whisker plot)** summarizes the **distribution** of numerical data. It shows the **median, quartiles, spread, and outliers**, making it useful for comparing datasets.

**Best used for:**

* **Box Plot:** Shows the distribution, spread, and outliers of data.

* **Line Plot:** Shows trends over time.
* **Bar Plot:** Compares categories.

## 8.2 Implementation

# ⬢ Level 9: Hexagonal Bin Plot (Hexbin Plot)

## 9.1 Theory of Hexbin Plot

A **hexagonal bin plot (hexbin plot)** groups nearby data points into **hexagonal bins** instead of plotting every point individually. It is useful for **large datasets** where scatter plots become overcrowded.

**Best used for:**

* 🚗 Car price vs. Mileage
* 🏠 House size vs. Price
* 🌍 Population density
* 📈 Large datasets with thousands of points

**Key Point:**

* **Scatter Plot:** Shows every individual data point (best for small datasets).
* **Hexbin Plot:** Groups points into hexagons to show data density (best for large datasets).

## 9.2 Making DataFrame

```python
rand_data = pd.DataFrame(np.random.rand(1000, 2), columns=["A", "B"])
rand_data
```

## 9.3 Implementation

```python
rand_data.plot.hexbin(
    x="A",
    y="B",
    gridsize=20,
    cmap="viridis"
)
```

# 📈 Level 10: KDE Plot (Kernel Density Estimation Plot)

## 10.1 Theory of KDE Plot

A **KDE plot** shows the **probability density (distribution)** of a numerical variable using a **smooth curve**. It is a smoother alternative to a histogram.

**Best used for:**

* 📚 Exam score distribution
* 💰 Salary distribution
* 🏠 House price distribution
* 📏 Height or weight distribution

**Key Point:**

* **Histogram:** Shows the distribution using bars.
* **KDE Plot:** Shows the same distribution using a smooth density curve, making patterns easier to see.

## 10.2 Implementation

```python
df.plot.kde()
```
