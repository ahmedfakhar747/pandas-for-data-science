# 📘 Level 3: Pandas Series — Working with 1D Data

## 3.1 What is Series?
Pandas Series is like a column iin a table, it is a one-dimensional array containing data of any type.

## 3.2 Conversion: LIST -> SERIES

```python
list1 = [10, 20, 30, 40, 50]
series1 = pd.Series(list1)
print(f"\nPandas series looks like this:\n{series1}")
```

## 3.3 Conversion: NUMPY ARRAY -> SERIES

```python
import numpy as np

array1 = np.array([100, 200, 300, 400, 500])
series2 = pd.Series(array1)
print(f"\nPandas series looks like this:\n{series2}")
```

## 3.4 Python List VS Pandas Series

| Feature        | List 🐍         | Series 🐼           |
| -------------- | --------------- | ------------------- |
| **Type**       | Python built-in | Pandas object       |
| **Index**      | Numeric only    | Custom or numeric   |
| **Operations** | Basic           | Advanced/vectorized |
| **Use**        | General purpose | Data analysis       |

👉 **List = simple container**
👉 **Series = labeled smart list** 🚀


## 3.5 Conversion: DICTIONARY -> SERIES

```python
dictionary1 = {
    "a": "Earth",
    "b": "Mars",
    "c": "Jupiter",
    "d": "Saturn"
}

series3 = pd.Series(dictionary1)
print(f"\nPython Dictionary to Pandas series:\n{series3}")
```

# 📝 Level 4: Advanced Pandas Series — Series Operations

## 4.1 Series from Label Lists
Our task is to take the data and label values as seperated lists and create Pandas Series

```python
labels = ["car1", "car2", "car3"]
cars = ["Honda", "Toyota", "BYD"]

car_series = pd.Series(data=cars, index=labels)
print(f"\nSeries with labels:\n{car_series}")
```
