# 📘 Level 3: Pandas Series — Working with 1D Data

## 3.1 What is Series?
Pandas Series is like a column iin a table, it is a one-dimensional array containing data of any type.

## 3.2 Conversion: LIST -> SERIES

```python
list1 = [10, 20, 30, 40, 50]
series1 = pd.Series(list1)
print(f"\nPandas series looks like this:\n{series1}")
```

## 3.3 Conversion: NUMPY ARRAY -> SERIES

```python
import numpy as np

array1 = np.array([100, 200, 300, 400, 500])
series2 = pd.Series(array1)
print(f"\nPandas series looks like this:\n{series2}")
```

## 3.4 Python List VS Pandas Series

| Feature        | List 🐍         | Series 🐼           |
| -------------- | --------------- | ------------------- |
| **Type**       | Python built-in | Pandas object       |
| **Index**      | Numeric only    | Custom or numeric   |
| **Operations** | Basic           | Advanced/vectorized |
| **Use**        | General purpose | Data analysis       |

👉 **List = simple container**
👉 **Series = labeled smart list** 🚀


## 3.5 Conversion: DICTIONARY -> SERIES

```python
dictionary1 = {
    "a": "Earth",
    "b": "Mars",
    "c": "Jupiter",
    "d": "Saturn"
}

series3 = pd.Series(dictionary1)
print(f"\nPython Dictionary to Pandas series:\n{series3}")
```

# 📝 Level 4: Intermediate Pandas Series — Series Indexing

## 4.1 Series from Label Lists
Our task is to take the data and label values as seperated lists and create Pandas Series

```python
labels = ["car1", "car2", "car3"]
cars = ["Honda", "Toyota", "BYD"]

car_series = pd.Series(data=cars, index=labels)
print(f"\nSeries with labels:\n{car_series}")
```

## 4.2 Accessing items in Pandas Series
We can access an element of Pandas Series by indexing it's label/index

```python
print(f"\nFirst element of `car_series`:\n{car_series["car1"]}")
print(f"\nSecond element of `car_series`:\n{car_series["car2"]}")
print(f"\nThird element of `car_series`:\n{car_series["car3"]}")
```

# 🔢 Level 5: Advanced Pandas Series — Series Operations

## 5.1 Pandas Series Addition
Pandas Series addition is element-wise addition based on matching labels.

```python
ser1 = pd.Series([50, 100, 150, 200], ["a", "b", "c", "d"])
ser2 = pd.Series([250, 300, 350, 400], ["b", "c", "d", "e"])

ser1 + ser2
```

## 5.2 Addition using add() method

```python
print(f"\nWith using add():\n{ser1.add(ser2)}")
```

When we use simple add() the result is same as direct addition, if we want to retain values of labels that does not match we can do like this:

```python
print(f"\nWith using `fill_value` argument:\n{ser1.add(ser2, fill_value=0)}")
```

## 5.3 Subtraction using sub() method

```python
print(f"\nUsing `fill_value` argument while subtracting:\n{ser2.sub(ser1, fill_value=0)}")
```

# 📑 Level 6: Python DataFrames— Mastering Tabular Data (2D Structure)
A **DataFrame** is a **2-dimensional data structure in Pandas** used to store and organize data in a tabular format with **rows and columns**, where each column can have a different data type.

* DataFrames are the workhouse of Pandas
