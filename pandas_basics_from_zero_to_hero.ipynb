# 📘 Level 3: Pandas Series — Working with 1D Data

## 3.1 What is Series?
Pandas Series is like a column iin a table, it is a one-dimensional array containing data of any type.

## 3.2 Conversion: LIST -> SERIES

```python
list1 = [10, 20, 30, 40, 50]
series1 = pd.Series(list1)
print(f"\nPandas series looks like this:\n{series1}")
```

## 3.3 Conversion: NUMPY ARRAY -> SERIES

```python
import numpy as np

array1 = np.array([100, 200, 300, 400, 500])
series2 = pd.Series(array1)
print(f"\nPandas series looks like this:\n{series2}")
```

## 3.4 Python List VS Pandas Series

| Feature        | List 🐍         | Series 🐼           |
| -------------- | --------------- | ------------------- |
| **Type**       | Python built-in | Pandas object       |
| **Index**      | Numeric only    | Custom or numeric   |
| **Operations** | Basic           | Advanced/vectorized |
| **Use**        | General purpose | Data analysis       |

👉 **List = simple container**
👉 **Series = labeled smart list** 🚀


## 3.5 Conversion: DICTIONARY -> SERIES

```python
dictionary1 = {
    "a": "Earth",
    "b": "Mars",
    "c": "Jupiter",
    "d": "Saturn"
}

series3 = pd.Series(dictionary1)
print(f"\nPython Dictionary to Pandas series:\n{series3}")
```

# 📝 Level 4: Intermediate Pandas Series — Series Indexing

## 4.1 Series from Label Lists
Our task is to take the data and label values as seperated lists and create Pandas Series

```python
labels = ["car1", "car2", "car3"]
cars = ["Honda", "Toyota", "BYD"]

car_series = pd.Series(data=cars, index=labels)
print(f"\nSeries with labels:\n{car_series}")
```

## 4.2 Accessing items in Pandas Series
We can access an element of Pandas Series by indexing it's label/index

```python
print(f"\nFirst element of `car_series`:\n{car_series["car1"]}")
print(f"\nSecond element of `car_series`:\n{car_series["car2"]}")
print(f"\nThird element of `car_series`:\n{car_series["car3"]}")
```

# 🔢 Level 5: Advanced Pandas Series — Series Operations

## 5.1 Pandas Series Addition
Pandas Series addition is element-wise addition based on matching labels.

```python
ser1 = pd.Series([50, 100, 150, 200], ["a", "b", "c", "d"])
ser2 = pd.Series([250, 300, 350, 400], ["b", "c", "d", "e"])

ser1 + ser2
```

## 5.2 Addition using add() method

```python
print(f"\nWith using add():\n{ser1.add(ser2)}")
```

When we use simple add() the result is same as direct addition, if we want to retain values of labels that does not match we can do like this:

```python
print(f"\nWith using `fill_value` argument:\n{ser1.add(ser2, fill_value=0)}")
```

## 5.3 Subtraction using sub() method

```python
print(f"\nUsing `fill_value` argument while subtracting:\n{ser2.sub(ser1, fill_value=0)}")
```

Explore how to **create, modify, and delete** columns in Pandas **DataFrames** to keep your data structured, flexible, and ready for analysis.

```python
data = pd.DataFrame(np.random.randn(4,6), columns=["A", "B", "C", "D", "E", "F"])
data
```

# 🔍 Level 8: DataFrame Indexing — Selecting Specific Data

We will start by creting a random DataFrame of 2x2

```python
randomData = pd.DataFrame(np.random.randint(100, 150, size=(3, 3)),
                          index=["2023 Sales", "2024 Sales", "2025 Sales"],
                          columns=["Product 1", "Product 2", "Product 3"])
randomData
```

## 8.1 Selecting Column
We discussed earlier that we can access column directly as below

```python
randomData["Product 1"]
```

## 8.2 Selecting Row
But we cannot select row **directly** from a DataFrame.

There are 2 methods to access row:
- **.loc[]** - uses label of row
- **.iloc[]** - uses index of row

### 8.2.1. Using .loc()

```python
# Using loc[] for aceesing sales across Years 2023, 2024, 2025
print(randomData.loc["2023 Sales"], "\n")
print(randomData.loc["2024 Sales"], "\n")
print(randomData.loc["2025 Sales"], "\n")
```

### 8.2.2 Using .iloc()

```python
# Using iloc[] for aceesing sales across Years 2023, 2024, 2025
print(randomData.iloc[0], "\n")
print(randomData.iloc[1], "\n")
print(randomData.iloc[2], "\n")
```

## 8.3 Selecting Particular Value

```python
# Column-first look up (Uses label names)
print(f"The Sales of product 1 in 2024 is: {randomData['Product 1']['2024 Sales']}")

# Label-based look up (df.loc[row_label, col_label])
print(f"Using loc[]: {randomData.loc["2024 Sales", "Product 1"]}")

# Position-based look up (df.iloc[row_index, col_index])
print(f"Using iloc[]: {randomData.iloc[1, 0]}")
```

## 8.4 Selecting Subset of DataFrame

```python
randomData.loc[["2023 Sales", "2024 Sales"], ["Product 2", "Product 3"]]
```

# 🎯 Level 9: DataFrame Conditional Selection — Isolating Targets
An important feature of Pandas is **Conditional Selection** using bracket notation which is **very similar** to NumPy

```python
randPopulation = pd.DataFrame(np.random.randint(10000000, 550000000, size=(3, 3)),
                          index=["2023 Population", "2024 Population", "2025 Population"],
                          columns=["Country 1", "Country 2", "Country 3"])
randPopulation
```

## 9.1 Checking Population Conditions

Here, we check which country populations are greater than **300,000,000**.  
The result will return **True** for matching values and **False** for non-matching values.

```python
randPopulation > 300000000
```

## 9.2 Filtering Countries by Population

Now, we use the condition to filter the DataFrame and display only the population values greater than **300,000,000**.


```python
randPopulation[randPopulation > 300000000]
```

## 9.3 Filtering Rows with Conditions

How to filter rows in a **DataFrame** using conditions.  
We check values inside a `column` and display only the rows that satisfy the **condition**.

```python
randPopulation[randPopulation["Country 1"] > 300000000]
```

## 9.4 Filtering Rows & Selecting a Column

Filter rows where `"Country 1"` population is greater than `300000000`, then display only the `"Country 1"` column.

```python
randPopulation[randPopulation["Country 1"] > 300000000]["Country 1"]
```

# 📈 Level 11: Data Analysis — Aggregation & Statistics

In this level, we dive into data analysis using Pandas aggregation and statistical functions. Through summaries, calculations, and descriptive statistics, we uncover meaningful patterns and insights hidden within the data, transforming raw information into valuable knowledge.

## 11.1 Basic Aggregation Functions
Using common aggregation methods to calculate summary statistics.


```python
print(iPhoneData["Price"].sum())
print(iPhoneData["Price"].mean())
print(iPhoneData["Price"].min())
print(iPhoneData["Price"].max())
print(iPhoneData["Price"].count())
```

## 11.2 Descriptive Statistics with `describe()`
Generate a statistical summary of numerical columns.

```python
iPhoneData.describe()
```

Provides:
- Count
- Mean
- Standard Deviation
- Minimum
- Quartiles (25%, 50%, 75%)
- Maximum

## 11.3 Finding Largest and Smallest Values
Identify top and bottom values in a dataset.

```python
print(iPhoneData["Price"].nlargest(5))
print(iPhoneData["Price"].nsmallest(5))
```

## 11.4 Counting Unique Values
Analyze the frequency of values in a column.

```python
iPhoneData["Country"].value_counts()
```

Find the number of unique values:

```python
iPhoneData["Country"].nunique()
```

## 11.5 Calculating Statistical Measures
Compute additional statistical metrics.

```python
print(iPhoneData["Price"].median())
print(iPhoneData["Price"].std())
print(iPhoneData["Price"].var())
```
