# 📊 Level 2: Pandas Correlation — Understanding Relationships

## 2.1 What is Correlation?

**Correlation** is a statistical concept used to measure the relationship between two variables or features. It helps us understand how one value changes compared to another. 📈

* A **positive correlation** means both variables increase together.
* A **negative correlation** means one increases while the other decreases.
* A correlation close to **0** means there is little or no relationship. 🎯

In Pandas, correlation is commonly used during **data analysis** to discover connections between numerical columns. 🐼


## 2.2 The `corr()` Method

Use the `corr()` method to calculate the correlation between **numerical columns** in a DataFrame.

* **Note:** Since the dataset contains non-numeric values, we will select only the numerical columns for correlation analysis. 📊

```python
corelation = titanic_data.select_dtypes(include="number").corr()
corelation
```

## 2.3 Good Corelation
A **good correlation** means two variables have a strong relationship with each other. 📈

* Correlation close to **+1** → strong positive relationship
* Correlation close to **-1** → strong negative relationship
* Correlation close to **0** → weak or no relationship

Example:

`0.92` means a very strong positive correlation. 🎯


## 2.4 Perfect Corelation
A **perfect correlation** occurs when two variables move exactly together in a predictable pattern. 🎯

* **+1** → perfect positive correlation 📈
* **-1** → perfect negative correlation 📉

Example:

`+1.0` means both variables increase together perfectly.


## 2.5 Quick Rule

| Correlation Value | Strength |
| ----------------- | -------- |
| `0.7 → 1.0`       | Strong   |
| `0.4 → 0.7`       | Moderate |
| `0.0 → 0.4`       | Weak     |

Same idea applies for negative values too.

# 🧹 Level 4: Data Cleaning — Preparing High-Quality Data

Data cleaning is the process of detecting and fixing incorrect, missing, or messy data in a dataset to improve data quality. 📊

### 🚫 Common Types of Bad Data

Bad data may include:

* ❌ Missing or empty values
* 🔁 Duplicate rows
* 📝 Incorrect data formats
* ⚠️ Invalid or wrong values


## 4.1 Loading a Messy Titanic Dataset

Load a real-world Titanic dataset containing missing values and inconsistent data for practicing **data cleaning** techniques in Pandas. 🧹📊

```python
titanic_dataset2 = pd.read_csv("/kaggle/input/datasets/heptapod/titanic/train_and_test2.csv")
titanic_dataset2
```

## 4.2 Handling Missing Values

How to identify and manage empty or missing cells in a DataFrame for cleaner and more reliable analysis.

### 4.2.1 Removing Columns Containing Only Zero Values

Identify columns where all values are `0` and remove them from the DataFrame to clean unnecessary data. 📊


```python
cols_to_drop = titanic_dataset2.columns[(titanic_dataset2 == 0).all()]
new_data = titanic_dataset2.drop(columns=cols_to_drop)
new_data
```

### 4.2.2 Adding Empty Columns (Handling Missing Features)

Sometimes we need placeholder columns while preparing or cleaning data. These empty columns help us simulate missing values and practice imputation techniques.

```python
new_data["Empty_1"] = None
new_data["Empty_2"] = None
new_data
```


If a column contains missing values but we want to assign a constant value (for example, when data is uniform or assumption-based), we can use `fillna()`:

```python
new_data.fillna({"Empty_1": 123}, inplace=True)
new_data
```

### 4.2.3 Handling Missing Values Using Mean Imputation

In real-world datasets, some values are known while others are missing. A common strategy is **mean imputation**, where missing values are replaced with the average of available data.

Here, we first simulate partial data in `Empty_2`:


```python
for i in range(0, 1309, 2):
    new_data.iloc[i, 10] = i + 2000
new_data
```

Now compute the mean of the existing values:

```python
meanValue = int(new_data["Empty_2"].mean())
meanValue
```

Finally, fill missing values using the computed mean:

```python
new_data.fillna({"Empty_2": meanValue}, inplace=True)
new_data
```

After this step, the dataset becomes complete with no missing gaps.

# 🤝 Level 6: DataFrame Combination — Merging, Joining & Concatenation

## 6.1 Sale DataFrames

For the sake of example we are taking two datasets for interpretation.

```python
import numpy as np

# Shared baseline data so the merge keys actually match
shared_products = ["Laptop", "Phone", "Tablet", "Headphones"]
shared_units = [45, 82, 18, 64]  # Fixed numbers so they match exactly

# Store 1 DataFrame
sales1 = {
    "Product": shared_products,
    "Units_Sold": shared_units,
    "Price_per_Unit": [1000, 800, 400, 150],
    "Total_Sales_Store1": [45000, 65600, 7200, 9600]
}
df_sales1 = pd.DataFrame(sales1)

# Store 2 DataFrame (Same products/units, different store performance metrics)
sales2 = {
    "Product": shared_products,
    "Units_Sold": shared_units,
    "Discount_Given": [50, 30, 10, 5],
    "Total_Sales_Store2": [42750, 63140, 7020, 9280]
}
df_sales2 = pd.DataFrame(sales2)
```

## 6.2 Merging DataFrames

The merge function in pandas facilitates the merging of datasets

### 6.2.1 Inner Merge

```python
inner_merged_sales = pd.merge(df_sales1, df_sales2, on=["Product", "Units_Sold"])
inner_merged_sales
```

### 6.2.2 Outer Merge

```python
outer_merged_sales = pd.merge(df_sales1, df_sales2, how="outer", on=["Product", "Units_Sold"])
outer_merged_sales
```

## 6.3 Joining DataFrames
Joining is a convenient method for combining the columns of two dataframes, it is same as merging but it is performed on indexes instead of keys

First we are making two new dataframes:

```python
products = ["Laptop", "Phone", "Tablet", "Headphones"]

# DataFrame 1: Base Pricing Data
df_prices = pd.DataFrame({
    "Units_Sold": [50, 85, 20, None],
    "Price_per_Unit": [None, 700, 400, 150]
}, index=products)

# DataFrame 2: Financial Performance Data
df_revenues = pd.DataFrame({
    "Total_Revenue": [50000, None, 8000, None],
    "Profit_Margin_%": [25, 30, None, 40]
}, index=products)
```

### 6.3.1 Inner Join

```python
df_prices.join(df_revenues)
```

### 6.3.1 Outer Join

```python
df_prices.join(df_revenues, how="outer")
```

## 6.4 Concatenation
Concatenation basically glues together dataframes. Dimensions should match along the axis we are concating on.

```python
df1 = pd.DataFrame({"A": ["A0", "A1", "A2"],
                   "B": ["B0", "B1", "B2"],
                   "C": ["C0", "C1", "C2"],},
                  index=[0, 1, 2])
df1
```

```python
df2 = pd.DataFrame({"A": ["A3", "A4", "A5"],
                   "B": ["B3", "B4", "B5"],
                   "C": ["C3", "C4", "C5"],},
                  index=[3, 4, 5])
df2
```

```python
pd.concat([df1, df2])
```

# 🧩 Level 11: Data Summarization — Creating Pivot Tables
A pivot table is a powerful data summarization tool used in data analysis (found in Excel, Google Sheets, Pandas, SQL, etc.). It allows you to take a large, messy dataset and instantly group, sort, count, total, or average it without writing complex formulas.

## 11.1 Initializing DataFrame
A pivot table instantly transforms this flat data into a clean summary grid:

```python
df = pd.DataFrame({
    "Order_ID": [1, 2, 3, 4, 5, 6],
    "Date": ["2026-07-01", "2026-07-01", "2026-07-02", "2026-07-02", "2026-07-03", "2026-07-03"],
    "Category": ["Clothing", "Electronics", "Clothing", "Electronics", "Clothing", "Electronics"],
    "Region": ["East", "West", "West", "East", "East", "West"],
    "Sales": [150, 1200, 200, 800, 300, 1500]
})
df
```

## 11.2 Using `.pivot_table()`
Pandas has a built-in `.pivot_table()` function that handles this automatically:

```python
pivot_df = df.pivot_table(
    values="Sales", 
    index="Category", 
    columns="Region", 
    aggfunc="sum"
)
pivot_df
```

# 🌐 Level 12: Web Data Extraction — Reading HTML Tables

## 12.1 The Syntax
We can pass a website URL, a local `.html` file path, or even a raw string of HTML code directly into the function:

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)"

headers = {'User-Agent': 'Mozilla/5.0'}
tables = pd.read_html(url, storage_options=headers)
```

## 12.2 Accessing Data

```python
# Check how many tables were found on the page
print(f"Total tables found: {len(tables)}")

# Get the first table as a proper DataFrame
df_population = tables[0]

# View the DataFrame
df_population.head()
```
