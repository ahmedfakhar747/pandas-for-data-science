# 📊 Level 2: Pandas Correlation — Understanding Relationships

## 2.1 What is Correlation?

**Correlation** is a statistical concept used to measure the relationship between two variables or features. It helps us understand how one value changes compared to another. 📈

* A **positive correlation** means both variables increase together.
* A **negative correlation** means one increases while the other decreases.
* A correlation close to **0** means there is little or no relationship. 🎯

In Pandas, correlation is commonly used during **data analysis** to discover connections between numerical columns. 🐼


## 2.2 The `corr()` Method

Use the `corr()` method to calculate the correlation between **numerical columns** in a DataFrame.

```python
studentData.select_dtypes(include="number").corr()
```
